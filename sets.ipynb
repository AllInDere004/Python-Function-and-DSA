```python
# set membership test
my_set=(1,2,3,4,5,6)
print(3 in my_set)
print(10 in my_set)
```

```python
# Mathematical operation
set1={1,2,3,4,5}
set2={4,5,6,7,8}

## Union
union_set=set1.union(set2)
print(union_set)

## intersection
intersect_set=set1.intersection(set2)
print(intersect_set)

set1.intersection_update(set2)
print(set1)
```

```python
## Difference
set1={1,2,3,4,5}
set2={4,5,6,7,8}
print(set1.difference(set2))
```

```python
# Symmetric Difference
set1.symmetric_difference(set2)
```

```python
# Sets method
set1={1,2,3,4,5}
set2={3,4,5}

# is subset
print(set1.issubset(set2))
# is superset
print(set1.issuperset(set2))
```

```python
# Mini Set function project
def unique_words(text):
    text=text.lower()
    for ch in ".,!?;:":
        text=text.replace(ch, "")
    words=text.split()
    return set(words)

text=input("Enter text: ")
uw=unique_words(text)

print("Unique Words: ",uw)
print("Count of unique words: ", len(uw))
```
