```python
# Excessing dictionaries elements
student={"name":"Soumalya","age":20,"grade":'A'}
print(student)
```

```python
# Accessing dictionaries elements
print(student['grade'])
print(student['age'])

# Accessing using get() method
print(student.get('grade'))
print(student.get('last_name'))
print(student.get('last_name','not Available'))
```

```python
# Modifying dictionary elements
# it is mutable, so you can add,update or delete elements
print(student)
```

```python
student["age"]=33       ## update a value
print(student)
student["address"]="India"      # added a new key and value
print(student)
```

```python
del student["grade"]        ## delete key and value pair
print(student)
```

```python
## Dictionaries methods

keys=student.keys()     # get all keys
print(keys)
values=student.values()     # get all values
print(values)

items=student.items()       # get all key values pair
print(items)
```

```python
# Shallow copy
student_copy=student
print(student)
print(student_copy)
```

```python
student_copy=student.copy()
print(student)
print(student_copy)
```

```python
student["name"]="Tamanya"
print(student)
print(student_copy)
```

```python
# Iterating over dictionaries
# we can  use loops for iterations over dictionaries,keys,values or items

# iterating over keys
for keys in student.keys():
    print(keys)
```

```python
# iterate over values
for values in student.values():
    print(values)

## iterate over key and values pair
for key,value in student.items():
    print(f"{key}:{value}")
```

```python
## Nested dictionaries
students={
    "Student1":{"Name":"Soumalya","age":20},
    "Student2":{"Name":"kanka","age":21}
}
print(students)
```

```python
# Access nested dictionaris elements
print(students["Student2"]["age"])
```

```python
# iterate over nested dictionaries
students.items()
```

```python
# iterate over nested dictionaries
for student_id,student_info in students.items():
    print(f"{student_id}:{student_info}")
    for key,value in student_info.items():
        print(f"{key}:{value}")
```

```python
# dictionary comprehension
squares={x:x**2 for x in range(5)}
print(squares)
```

```python
# conditional dictionary comprehension
even={x:x*2 for x in range(10) if x%2==0}
print(even)
```

```python
# Practical examples
# Use a dictionary to count the frequency of elements in list
number=[1,2,3,4,2,3,3,4,4,4]
frequency={}
for num in number:
    if num in frequency:
        frequency[num]+=1
    else:
        frequency[num]=1
print(frequency)
```

```python
# Merge tw=o ductionaries into one
dict1={"a":1,"b":2}
dict2={"b":3,"c":5}
merge_dict={**dict1,**dict2}
print(merge_dict)
```

## Small mini Project

```python
text=input("Enter a text: ")

text=text.lower()

for ch in ",.?!:;":
    text=text.replace(ch, "")

words=text.split()

file={}

for word in words:
    if word in file:
        file[word] += 1
    else:
        file[word] = 1

print("\n---Word Frequency---")
for k,v in file.items():
    print(k, ":", v)
```
