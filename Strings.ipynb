```python
# Mini Project: Text Analyzer
text=input("Enter text: ")
def character_count(s):
    return len(s)

def Word_count(s):
    words=s.split()
    return len(words)

def Vowel_count(s):
    count=0
    for ch in s.lower():
        if ch in "aeiou":
            count+=1
    return count

def upper_lower(s):
    upper=0
    lower=0
    for ch in s:
        if ch.isupper():
            upper+=1
        elif ch.islower():
            lower+=1
    return upper,lower

def reverse_str(s):
    return s[::-1]

print("\n--- Text Analysis ---\n")
print("Text : ",text)
print("Characters: ", character_count(text))
print("Words: ",Word_count(text))
print("Vowels: ",Vowel_count(text))

u, l= upper_lower(text)

print("UPPERCASE LETTERS: ",u)
print("lowercase letters: ",l)

print("Reversed text: ",reverse_str(text))
```
