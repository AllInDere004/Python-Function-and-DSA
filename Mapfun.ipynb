## map() function
##### this function applies a given function to all items in an input list and returns a map object. This is particularly useful for transforming data in a list comprehensively

```python
def square(x):
    return x*x
square(10)
```

```python
numbers=[1,2,3,4,5,6]
list(map(square,numbers))
```

```python
## lambda func with map
numbers=[1,2,3,4,5,6]
list(map(lambda x:x*x,numbers))
```

```python
## map multiple iterables

numbers1=[1,2,3]
numbers2=[4,5,6]

add_num=list(map(lambda x,y:x+y,numbers1,numbers2))
print(add_num)
```

```python
chr_number=['1','2','3','4','5']
int_number=list(map(int,chr_number))

print(int_number)
```

```python
fruit=["apple","banana","mango","guava","cherry"]
upper_str=list(map(str.upper,fruit))
print(upper_str)
```

```python
def get_name(person):
    return person['name']

people=[
    {'name':"Soumalya",'age':25},
    {'name':"Kanka",'age':20}
]
list(map(get_name,people))
```
