```python
# Packing and Unpacking of Tuple
#packing
packes_tuple=1,"Helo",3.142
print(packes_tuple)
```

```python
# unpacking
a,b,c=packes_tuple
print(a)
print(b)
print(c)
```

```python
# Unpacking with *
number  
first,*middle,last=number
print(first)
print(middle)
print(last)
```

```python
# Nested tuples
# Nested List
list=[[1,2,3,4],[5,6,7,8],[1,"Hello",3.14,"c"]]
print(list[0])
print(list[1][2])
print(list[2][0:3])
```

```python
list=[[1,2,3,4],[5,6,7,8],(1,"Hello",3.14,"c")]
print(list[2][0:3])
```

```python
# nested tuples
nested_tuples=((1,2,3),("a","b","c"),(True,False))

# access the elements inside the tuple
print(nested_tuples[1][2])
```

```python
# iterating over nested tuples
for sub_tuples in nested_tuples:
    for items in sub_tuples:
        print(items,end=" ")
    print()
```
