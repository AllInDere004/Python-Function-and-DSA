##### List Comprehension

Basic Syntax [expression for item in iterable]

with conditional logic [expression for item in iterable if condition]

Nested List comprehension [expression for item1 in iterable1 for item2 in iterable2]



```python
#Basic List comprehension

square=[num**2 for num in range(10)]
print(square)
```

```python
# List comprehension with condition
[num for num in range(10) if num%2==0]
```

```python
# Nested List Comprehension

list1=[1,2,3,4]
list2=['a','b','c','d']

pair=[[i,j] for i in list1 for j in list2]
print(pair)
```

```python
# List  comprehension with function call
words=["banana","apple","comprehension"]
lengths=[len(word) for word in words]

print(lengths)
```

#### Create a To Do List **


```python
to_do_list=["Buy grocerries","Clean the house","Pay bills"]

# Adding to task
to_do_list.append("Schedule meeting")
to_do_list.append("Go for a run")

# Remove a complete task
to_do_list.remove("Clean the house")

# Check if a task is in the list
if "Pay bills" in to_do_list:
    print("Dont forget to pay the utility bills")

print("To do list remaining")
for task in to_do_list:
    print(f"-{task}")
```
